### Step 4 : Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cat
```

```python
for i in cat:
    df[i].value_counts().plot(kind='bar', 
                              title=f'Countplot for {i}',
                              figsize=(16,8))
    plt.show()
```

```python
con
```

```python
for i in con:
    plt.figure(figsize=(16,8))
    sns.histplot(data=df, x=i, kde=True)
    plt.title(f'Histogram for {i}')    
    plt.show()
```

### KDE - Kernel Density Estimate

### Bivariate Analysis - Analysis of 2 variables togethor

```python
con
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x='Weight', y='MPG.city')
plt.title('Scatterplot for Weight vs MPG.city')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x='Length', y='Price')
plt.title('Scatterplot for Length vs Price')
plt.grid()
plt.show()
```

### Correlation heatmap

```python
df[con].corr()
```

```python
plt.figure(figsize=(25, 25))
sns.heatmap(df[con].corr(), annot=True)
plt.show()
```
